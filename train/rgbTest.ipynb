```python
import os
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def load_image(fileName):
    return mpimg.imread(fileName)

def display_image(fileName):
    imgPlot = plt.imshow(fileName)
    plt.show()

def feature_extract(imgPatch):

    mean = np.mean(imgPatch, axis=(0,1))
    std = np.std(imgPatch, axis=(0,1))
    
    print("Mean: ", mean)
    print("Standard Deviation: ", std)
```

```python
tram = load_image("tram.png")
display_image(tram)
feature_extract(tram)
```

```python
tram = load_image("half.png")
display_image(tram)
feature_extract(tram)
```

```python
tram = load_image("building.png")
display_image(tram)
feature_extract(tram)
```
